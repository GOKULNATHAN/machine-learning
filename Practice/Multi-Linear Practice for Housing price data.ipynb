#### Importing the Housing data CSV by considering that there are no headers in the CSV and also considering column separator is whitespace

```python
housing_data = pd.read_csv(r"C:\Learning\Data\housing.data",header=None,delim_whitespace=True)
```

#### Adding Column names to the data frame

```python
housing_data.columns = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
```

#### Declaring list of input variables and target/dependant variable

```python
x_columns = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT"]
pdx = housing_data[x_columns]
pdy = housing_data["MEDV"]
```

#### Split the training data and test data by considering 70:30 ratio

```python
x_train, x_test,y_train,y_test = train_test_split(pdx,pdy,train_size=70/100,random_state = 70)
```

#### Intercepting the data for Ordinary Least Square Model

```python
x_train_cons = sm.add_constant (x_train)
x_test_cons = sm.add_constant (x_test)
full_mod = sm.OLS(y_train,x_train_cons)
```

#### Fitting the model and Printing the Result

```python
regression_full = full_mod.fit()
print(regression_full.summary())
```

```python
print("Variance Inflation Factor")
cnames = x_train.columns
for i in (np.arange(0,len(cnames))):
    xvars = list(cnames)
    yvar = xvars.pop(i)
    mod = sm.OLS(x_train[yvar],sm.add_constant(x_train_cons[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    print(yvar,round(vif,3))
```

#### In the above iteration R-Square Value is 0.738 and Adjacent R-Square Value is 0.728. 
#### Probability value is high for the columns AGE - 0.659 & INDUS - 0.451
#### Going to Run one more iteration by removing the independant variable AGE

```python
x_columns = ["CRIM","ZN","INDUS","CHAS","NOX","RM","DIS","RAD","TAX","PTRATIO","B","LSTAT"]
pdx = housing_data[x_columns]
pdy = housing_data["MEDV"]

x_train, x_test,y_train,y_test = train_test_split(pdx,pdy,train_size=70/100,random_state = 70)

x_train_cons = sm.add_constant (x_train)
x_test_cons = sm.add_constant (x_test)
full_mod = sm.OLS(y_train,x_train_cons)

regression_full = full_mod.fit()
print("Iteration-2", regression_full.summary())

print("Variance Inflation Factor")
cnames = x_train.columns
for i in (np.arange(0,len(cnames))):
    xvars = list(cnames)
    yvar = xvars.pop(i)
    mod = sm.OLS(x_train[yvar],sm.add_constant(x_train_cons[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    print(yvar,round(vif,3))
```

#### In the 2nd Iteration, R-Square Value is 0.738 and Adjacent R-Square Value is 0.729. There is not that much difference.

#### In the 3rd iteration too, there is not that much difference. R-Square Value is 0.738 and Adjacent R-Square Value is 0.729
